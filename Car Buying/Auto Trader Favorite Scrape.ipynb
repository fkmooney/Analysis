```python
import pandas as pd
import bs4
```

```python
#in order to get list of cars, open car list in browser, open page source, copy HTML, and save to text file

f = "autotrader-2022-11-04.txt"
f = open(f, encoding="mbcs")
soup = bs4.BeautifulSoup(f, 'html.parser')
```

```python
container = soup.find("div", {"class": "container"})
```

```python
inventory = container.find("div", {"data-cmp": "inventoryListing"})
for car in inventory:
    print(car.text)
```

```python
inventory = [car.text for car in container.findAll("div", {"data-cmp":"inventoryListing"})] 
df = pd.DataFrame(inventory)
```

```python
df2 = df[~df.iloc[:,0].str.startswith('Save')]
df2
```

```python
df2[['Nothing', 'Full']]= df2[0].str.split('Share', expand=True)
df2[['Full', 'Gas']] = df2['Full'].str.split('miles', expand=True)
df2[['Mileage', 'Dealer']] = df2['Gas'].str.split('Highway', expand=True)
#df2[['Dealer', 'link', 'link2']] = df2['Dealer'].str.split('Get', expand=True)
df2[['Dealer', 'link',]] = df2['Dealer'].str.split('Get', expand=True)
df2 = df2.drop('Nothing', axis=1)
df2 = df2.drop('Gas', axis=1)
df2 = df2.drop(0, axis=1)
df2
```

```python
df2['Odometer'] = df2['Full'].str[-7:]
df2['Full'] = df2['Full'].str[:-7]
df2['Price'] = df2['Full'].str[-6:]
df2['Full'] = df2['Full'].str[:-6]
df2
```

```python
df2.to_csv('autotrader_scrape.csv')
```
